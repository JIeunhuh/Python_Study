## __Matplotlib__
  
  - 파이썬에서 데이터 시각화를 위한 일반적인 라이브러리
  - numpy와 함께 사용
  - 다양한 유형의 그래프 그리기 가능

```python
import matplotlib.pyplot as plt
import numpy as np

# 기본 그래프

# x,y data 생성
x = np.arange(0,10,0.1)
y = np.sin(x)

# 선 그래프(그래프 스타일 변경 가능)
plt.plot(x,y,color='red',linestyle='--',linewidth=2)

# 그래프 타이틀과 축 라벨 설정
plt.title('Sine graph')
plt.xlabel('x')
plt.ylabel('y')

# 그래프 출력
plt.show()
```
